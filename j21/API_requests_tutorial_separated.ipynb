# آموزش کار با API در پایتون و کتابخانه requests```pythonimport requests

# ارسال درخواست GET به یک URL
response = requests.get('https://api.example.com/data')

# بررسی وضعیت درخواست
if response.status_code == 200:
    print('درخواست موفقیت‌آمیز بود!')
    print('داده‌ها:', response.json())  # تبدیل داده‌های JSON به دیکشنری پایتون
else:
    print('خطا در درخواست:', response.status_code)```در این بخش به شما نحوه کار با API‌ها و استفاده از کتابخانه `requests` را در پایتون آموزش می‌دهیم.```pythonimport requests

# داده‌هایی که می‌خواهیم ارسال کنیم
data = {
    'name': 'John Doe',
    'age': 30
}

# ارسال درخواست POST به URL
response = requests.post('https://api.example.com/user', data=data)

# بررسی وضعیت درخواست
if response.status_code == 201:
    print('کاربر جدید با موفقیت ایجاد شد!')
    print('داده‌های بازگشتی:', response.json())  # نمایش پاسخ سرور
else:
    print('خطا در درخواست:', response.status_code)```### 1. آشنایی با API```pythonimport requests

# هدرهایی که می‌خواهیم ارسال کنیم
headers = {
    'Authorization': 'Bearer your_token_here'
}

# ارسال درخواست GET با هدر
response = requests.get('https://api.example.com/protected-data', headers=headers)

if response.status_code == 200:
    print('داده‌های محافظت‌شده:', response.json())
else:
    print('خطا در درخواست:', response.status_code)```### 2. نصب کتابخانه `requests````pythonimport requests

try:
    response = requests.get('https://api.example.com/data')
    response.raise_for_status()  # بررسی اگر خطای HTTP وجود داشته باشد
    print('داده‌ها:', response.json())
except requests.exceptions.RequestException as e:
    print('خطا در ارسال درخواست:', e)```برای شروع، باید کتابخانه `requests` را نصب کنید. این کار را با استفاده از pip می‌توانید انجام دهید:```pythonimport requests

params = {
    'search': 'python',
    'page': 2
}

response = requests.get('https://api.example.com/search', params=params)

if response.status_code == 200:
    print('نتایج جستجو:', response.json())
else:
    print('خطا در درخواست:', response.status_code)``````bash```pythonimport requests
