```python
import matplotlib.pyplot as plt
```

```python
# داده‌ها
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# رسم نمودار
plt.title('نمودار خطی ساده')
plt.xlabel('محور X')
plt.ylabel('محور Y')
plt.plot(x, y, color='red', linestyle='--', marker='x')

# نمایش نمودار
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 9]

# رسم نمودار ستونی
plt.bar(categories, values)

# نمایش نمودار
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 4, 5, 6, 6, 6, 7, 8, 9]

# رسم هیستوگرام
plt.hist(data, bins=10)

# نمایش نمودار
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# رسم نمودار دایره‌ای
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# نمایش نمودار
plt.show()
plt.savefig('p1.png')
```

```python
fig, axes = plt.subplots(1, 2)  # 1 ردیف و 2 ستون

axes[0].plot(x, y)
axes[0].set_title('نمودار 1')

axes[1].bar(categories, values)
axes[1].set_title('نمودار 2')

# نمایش نمودارها
plt.show()
```
