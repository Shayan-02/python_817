```python

import random
import statistics

def simulate_processing(num_parts, part_type_probs, part_type_means, part_type_stdevs, arrival_probs, arrival_intervals):
    total_time = 0
    processed_parts = 0

    while processed_parts < num_parts:
        part_type = random.choices(range(len(part_type_probs)), part_type_probs)[0]
        arrival_prob = random.random()

        if arrival_prob < arrival_probs[part_type]:
            part_processing_time = random.normalvariate(part_type_means[part_type], part_type_stdevs[part_type])
            total_time += part_processing_time
            processed_parts += 1

        arrival_interval = random.choices(arrival_intervals, arrival_probs)[0]
        total_time += arrival_interval

    return total_time

# Parameters
num_parts = 10
part_type_probs = [0.5, 0.3, 0.2]
part_type_means = [30, 40, 50]
part_type_stdevs = [3, 4, 7]
arrival_probs = [0.3, 0.3, 0.4]
arrival_intervals = [15, 25, 35]

# Simulate single machine
single_machine_time = simulate_processing(num_parts, part_type_probs, part_type_means, part_type_stdevs, arrival_probs, arrival_intervals)

# Simulate two machines in parallel
parallel_machines_time = simulate_processing(num_parts, part_type_probs, part_type_means, part_type_stdevs, arrival_probs, arrival_intervals) / 2

# Simulate three machines in parallel
parallel_machines_time_3 = simulate_processing(num_parts, part_type_probs, part_type_means, part_type_stdevs, arrival_probs, arrival_intervals) / 3

print("Single Machine Time:", single_machine_time)
print("Parallel Machines (2) Time:", parallel_machines_time)
print("Parallel Machines (3) Time:", parallel_machines_time_3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_production(num_machines):
    # تعریف پارامترهای شبیه‌سازی
    num_iterations = 1000  # تعداد تکرار شبیه‌سازی
    production_time = 30  # زمان تولید (به ثانیه)
    arrival_times = [10, 20, 30]  # زمان‌های ورود قطعات
    arrival_probs = [0.3, 0.3, 0.4]  # احتمال ورود قطعات

    # تعریف متغیرهای برای ذخیره نتایج
    delays = []  # لیست زمان تاخیر قطعات
    wips = []  # لیست تعداد قطعات در حال پردازش
    total_parts = []  # لیست تعداد کل قطعات پردازش شده

    for _ in range(num_iterations):
        # تعریف متغیرهای برای هر تکرار شبیه‌سازی
        parts_in_progress = 0  # تعداد قطعات در حال پردازش
        total_processed_parts = 0  # تعداد کل قطعات پردازش شده
        current_time = 0  # زمان فعلی

        for _ in range(production_time):
            # تولید قطعه به صورت تصادفی بر اساس توزیع ورود قطعات
            arrival = np.random.choice(arrival_times, p=arrival_probs)
            if current_time % arrival == 0:
                parts_in_progress += 1

            # انجام کار روی قطعات
            if parts_in_progress > 0:
                parts_in_progress -= num_machines
                if parts_in_progress < 0:
                    parts_in_progress = 0
                    total_processed_parts += num_machines
                else:
                    total_processed_parts += num_machines

            # افزایش زمان
            current_time += 1

        # محاسبه زمان تاخیر قطعات
        delay = total_processed_parts * arrival_times[0] - production_time
        delays.append(delay)

        # ذخیره تعداد قطعات در حال پردازش و کل قطعات پردازش شده
        wips.append(parts_in_progress)
        total_parts.append(total_processed_parts)

    return delays, wips, total_parts


# شبیه‌سازی با یک ماشین
delays_1, wips_1, total_parts_1 = simulate_production(1)
# شبیه‌سازی با دو ماشین
delays_2, wips_2, total_parts_2 = simulate_production(2)
# شبیه‌سازی با سه ماشین
delays_3, wips_3, total_parts_3 = simulate_production(3)

# رسم نمودار زمان تاخیر
plt.plot(range(len(delays_1)), delays_1, label='1 Machine')
plt.plot(range(len(delays_2)), delays_2, label='2 Machines')
plt.plot(range(len(delays_3)), delays_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Delay Time')
plt.title('Average Delay Time Comparison')
plt.legend()
plt.show()

# رسم نمودار WIP
plt.plot(range(len(wips_1)), wips_1, label='1 Machine')
plt.plot(range(len(wips_2)), wips_2, label='2 Machines')
plt.plot(range(len(wips_3)), wips_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Work in Progress (WIP)')
plt.title('WIP Comparison')
plt.legend()
plt.show()

# رسم نمودار کل قطعات پردازش شده
plt.plot(range(len(total_parts_1)), total_parts_1, label='1 Machine')
plt.plot(range(len(total_parts_2)), total_parts_2, label='2 Machines')
plt.plot(range(len(total_parts_3)), total_parts_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Total Processed Parts')
plt.title('Total Processed Parts Comparison')
plt.legend()
plt.show()
```

```python
import random

def simulate_production(num_machines):
    delays = []
    wips = []
    total_parts = []

    # دستگاه‌ها و پارامترهای توزیع تصادفی
    machines = ['Machine 1', 'Machine 2', 'Machine 3']
    machine_times = {'Machine 1': (10, 20), 'Machine 2': (20, 30), 'Machine 3': (30, 40)}
    part_types = ['Type A', 'Type B', 'Type C']
    part_distributions = {'Type A': (0.5, 30, 3), 'Type B': (0.3, 40, 4), 'Type C': (0.2, 50, 7)}

    parts_in_progress = 0
    total_processed_parts = 0

    for i in range(100):  # تعداد تکرار شبیه‌سازی

        # محاسبه زمان تاخیر ورود قطعه
        arrival_time = random.randint(machine_times[machines[0]][0], machine_times[machines[0]][1])

        # انتخاب نوع قطعه به صورت تصادفی با توجه به درصد هر نوع
        part_type = random.choices(part_types, [part_distributions[part][0] for part in part_types])[0]

        # محاسبه زمان تولید قطعه با توجه به توزیع نرمال
        production_time = random.gauss(part_distributions[part_type][1], part_distributions[part_type][2])

        # محاسبه زمان تاخیر
        delay = max(0, arrival_time - production_time)

        # بروزرسانی تعداد قطعات در حال پردازش
        parts_in_progress += 1

        # بروزرسانی تعداد قطعات پردازش شده
        total_processed_parts += 1

        # ذخیره زمان تاخیر و تعداد قطعات در حال پردازش و کل قطعات
        delays.append(delay)
        wips.append(parts_in_progress)
        total_parts.append(total_processed_parts)

    return delays, wips, total_parts

# شبیه‌سازی عملکرد یک ماشین
delays_1, wips_1, total_parts_1 = simulate_production(1)

# شبیه‌سازی عملکرد دو ماشین موازی
delays_2, wips_2, total_parts_2 = simulate_production(2)

# شبیه‌سازی عملکرد سه ماشین موازی
delays_3, wips_3, total_parts_3 = simulate_production(3)

# رسم نمودارها
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(delays_1, label='1 Machine')
plt.plot(delays_2, label='2 Machines')
plt.plot(delays_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Delay (seconds)')
plt.title('Delays')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(wips_1, label='1 Machine')
plt.plot(wips_2, label='2 Machines')
plt.plot(wips_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('WIP (Work in Progress)')
plt.title('Work in Progress')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(total_parts_1, label='1 Machine')
plt.plot(total_parts_2, label='2 Machines')
plt.plot(total_parts_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Total Processed Parts')
plt.title('Total Processed Parts')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def simulate_production(num_machines):
    delays = []
    wips = []
    total_parts = []

    # دستگاه‌ها و پارامترهای توزیع تصادفی
    machines = ['Machine 1', 'Machine 2', 'Machine 3']
    machine_times = {'Machine 1': (10, 20), 'Machine 2': (20, 30), 'Machine 3': (30, 40)}
    part_types = ['Type A', 'Type B', 'Type C']
    part_distributions = {'Type A': (0.5, 30, 3), 'Type B': (0.3, 40, 4), 'Type C': (0.2, 50, 7)}

    parts_in_progress = 0
    total_processed_parts = 0
    time_elapsed = 0

    while total_processed_parts < 100:
        # محاسبه زمان تاخیر ورود قطعه
        arrival_time = random.uniform(10, 40)

        # انتخاب نوع قطعه به صورت تصادفی با توجه به درصد هر نوع
        part_type = random.choices(part_types, [part_distributions[part][0] for part in part_types])[0]

        # محاسبه زمان تولید قطعه با توجه به توزیع نرمال
        production_time = random.gauss(part_distributions[part_type][1], part_distributions[part_type][2])

        # محاسبه زمان تاخیر
        delay = max(0, arrival_time - time_elapsed)

        # بروزرسانی زمان گذشته شده
        time_elapsed += delay

        # بروزرسانی تعداد قطعات در حال پردازش
        parts_in_progress += 1

        # بروزرسانی تعداد قطعات پردازش شده
        total_processed_parts += 1

        # ذخیره زمان تاخیر و تعداد قطعات در حال پردازش و کل قطعات
        delays.append(delay)
        wips.append(parts_in_progress)
        total_parts.append(total_processed_parts)

    return delays, wips, total_parts

# شبیه‌سازی عملکرد یک ماشین
delays_1, wips_1, total_parts_1 = simulate_production(1)

# شبیه‌سازی عملکرد دو ماشین موازی
delays_2, wips_2, total_parts_2 = simulate_production(2)

# شبیه‌سازی عملکرد سه ماشین موازی
delays_3, wips_3, total_parts_3 = simulate_production(3)

# رسم نمودارها
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(delays_1, label='1 Machine')
plt.plot(delays_2, label='2 Machines')
plt.plot(delays_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Delay (seconds)')
plt.title('Delays')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(wips_1, label='1 Machine')
plt.plot(wips_2, label='2 Machines')
plt.plot(wips_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('WIP (Work in Progress)')
plt.title('Work in Progress')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(total_parts_1, label='1 Machine')
plt.plot(total_parts_2, label='2 Machines')
plt.plot(total_parts_3, label='3 Machines')
plt.xlabel('Iteration')
plt.ylabel('Total Processed Parts')
plt.title('Total Processed Parts')
plt.legend()

plt.tight_layout()
plt.show()
```
